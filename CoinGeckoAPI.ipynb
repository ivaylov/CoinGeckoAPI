```python
import matplotlib.pyplot as plt
from datetime import datetime

from pycoingecko import CoinGeckoAPI
cg = CoinGeckoAPI()
```

```python
cg.get_price(ids='shiba-inu,ethereum,bitcoin',vs_currencies='usd')
```

```python
btc = cg.get_coin_market_chart_by_id(id='bitcoin', vs_currency='usd', days=1)
eth = cg.get_coin_market_chart_by_id(id='ethereum', vs_currency='usd', days=1)
shib = cg.get_coin_market_chart_by_id(id='shiba-inu', vs_currency='usd', days=1)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(8,6)
ax.plot([datetime.utcfromtimestamp(x[0]/1000) for x in btc['prices']], 
        [x[1] for x in btc['prices']], label='btc')
ax.set_xlabel('Hours')
ax.set_ylabel('Price *US$)')
ax.legend();
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(8,6)
ax.plot([datetime.utcfromtimestamp(x[0]/1000) for x in eth['prices'] ], 
        [x[1] for x in eth['prices']], label='eth')
ax.set_xlabel('Hours')
ax.set_ylabel('Price *US$)')
ax.legend();
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(8,6)
ax.plot([datetime.utcfromtimestamp(x[0]/1000) for x in shib['prices'] ], 
        [x[1] for x in shib['prices']], label='shib')
ax.set_xlabel('Hours')
ax.set_ylabel('Price *US$)')
ax.legend();
```
